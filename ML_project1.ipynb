```python
# Restrict ingredient changes
low  = orig_ing * 0.50
high = orig_ing.copy()     # <-- do not allow increases

egg_idx = ingredient_columns.index("Egg_Count")
low[egg_idx] = max(1, int(low[egg_idx]))

best_cost = total_cake_cost(orig_ing)
best_ing = orig_ing.copy()
best_quality_encoded = orig_quality_encoded

quality_order = {"Poor":1, "Average":2, "Good":3, "Excellent":4}
orig_label = lab.inverse_transform([orig_quality_encoded])[0]
orig_rank = quality_order[orig_label]

for i in range(500):   # more iterations

    cand = np.random.uniform(low, high)
    cand[egg_idx] = max(1, int(round(cand[egg_idx])))

    cand_full = sample_full.copy()
    for j, col in enumerate(ingredient_columns):
        cand_full[col] = cand[j]

    cand_scaled = scaler.transform(pd.DataFrame([cand_full], columns=full_feature_columns))
    pred_encoded = model.predict(cand_scaled)[0]
    pred_label = lab.inverse_transform([pred_encoded])[0]
    pred_rank = quality_order[pred_label]

    # must keep same or better quality
    if pred_rank < orig_rank:
        continue

    cand_cost = total_cake_cost(cand)

    if cand_cost < best_cost:
        best_cost = cand_cost
        best_ing = cand.copy()
        best_quality_encoded = pred_encoded
```

```python
cost_saving = round(orig_cost - best_cost, 2)
if cost_saving < 0:
    cost_saving = 0
    best_ing = orig_ing.copy()
    best_quality_encoded = orig_quality_encoded
```

```python

import pickle
df = pd.read_csv("cleaned_cake_dataset.csv")
FEATURE_COLS = [
    "Cake_Flavor",
    "Flour_Quantity (g)",
    "Sugar_Content (g)",
    "Butter_Quantity (g)",
    "Egg_Count",
    "Milk_Quantity (ml)"
]

TARGET = "Cake_Quality"

df_model = df[FEATURE_COLS + [TARGET]].copy()

def clean_numeric(col):
    return (
        df_model[col]
        .astype(str)
        .str.replace("[^0-9.]", "", regex=True)
        .replace("", np.nan)
        .astype(float)
    )

for col in [
    "Flour_Quantity (g)",
    "Sugar_Content (g)",
    "Butter_Quantity (g)",
    "Milk_Quantity (ml)"
]:
    df_model[col] = clean_numeric(col)

df_model.dropna(inplace=True)
flavor_encoder = LabelEncoder()
df_model["Cake_Flavor"] = flavor_encoder.fit_transform(df_model["Cake_Flavor"])

quality_encoder = LabelEncoder()
df_model["Cake_Quality"] = quality_encoder.fit_transform(df_model["Cake_Quality"])

INGREDIENT_COLS = [
    "Flour_Quantity (g)",
    "Sugar_Content (g)",
    "Butter_Quantity (g)",
    "Egg_Count",
    "Milk_Quantity (ml)"
]


scaler = StandardScaler()
df_model[INGREDIENT_COLS] = scaler.fit_transform(df_model[INGREDIENT_COLS])

X = df_model[FEATURE_COLS]
y = df_model[TARGET]

model = RandomForestClassifier(
    n_estimators=300,
    max_depth=12,
    random_state=42
)

model.fit(X, y)

with open("model.pkl", "wb") as f:
    pickle.dump(model, f)

with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

with open("label_encoder.pkl", "wb") as f:
    pickle.dump(quality_encoder, f)

with open("ingredient_columns.pkl", "wb") as f:
    pickle.dump(INGREDIENT_COLS, f)

with open("feature_columns.pkl", "wb") as f:
    pickle.dump(FEATURE_COLS, f)

with open("flavor_encoder.pkl", "wb") as f:
    pickle.dump(flavor_encoder, f)

```
